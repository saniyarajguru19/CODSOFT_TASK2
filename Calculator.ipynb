```python
import tkinter as tk
from tkinter import messagebox

# Function to perform addition
def add():
    try:
        result = float(entry_num1.get()) + float(entry_num2.get())
        result_label.config(text="Result: " + str(result))
    except ValueError:
        messagebox.showerror("Error", "Please enter valid numbers.")

# Function to perform subtraction
def subtract():
    try:
        result = float(entry_num1.get()) - float(entry_num2.get())
        result_label.config(text="Result: " + str(result))
    except ValueError:
        messagebox.showerror("Error", "Please enter valid numbers.")

# Function to perform multiplication
def multiply():
    try:
        result = float(entry_num1.get()) * float(entry_num2.get())
        result_label.config(text="Result: " + str(result))
    except ValueError:
        messagebox.showerror("Error", "Please enter valid numbers.")

# Function to perform division
def divide():
    try:
        num2 = float(entry_num2.get())
        if num2 == 0:
            messagebox.showerror("Error", "Cannot divide by zero.")
        else:
            result = float(entry_num1.get()) / num2
            result_label.config(text="Result: " + str(result))
    except ValueError:
        messagebox.showerror("Error", "Please enter valid numbers.")

# Create the main window
root = tk.Tk()
root.title("Simple Calculator")

# Set background color
root.configure(bg="#f1c40f")

# Create labels
label_num1 = tk.Label(root, text="Enter first number:", bg="#f1c40f")
label_num1.grid(row=0, column=0, padx=5, pady=5, sticky="w")

label_num2 = tk.Label(root, text="Enter second number:", bg="#f1c40f")
label_num2.grid(row=1, column=0, padx=5, pady=5, sticky="w")

result_label = tk.Label(root, text="Result:", bg="#f1c40f")
result_label.grid(row=4, column=0, columnspan=2, padx=5, pady=5, sticky="w")

# Create entry fields
entry_num1 = tk.Entry(root, bg="#ecf0f1")
entry_num1.grid(row=0, column=1, padx=5, pady=5)

entry_num2 = tk.Entry(root, bg="#ecf0f1")
entry_num2.grid(row=1, column=1, padx=5, pady=5)

# Create buttons
button_add = tk.Button(root, text="Add", command=add, bg="#2ecc71", fg="#000000")
button_add.grid(row=2, column=0, padx=5, pady=5)

button_subtract = tk.Button(root, text="Subtract", command=subtract, bg="#3498db", fg="#000000")
button_subtract.grid(row=2, column=1, padx=5, pady=5)

button_multiply = tk.Button(root, text="Multiply", command=multiply, bg="#e74c3c", fg="#000000")
button_multiply.grid(row=3, column=0, padx=5, pady=5)

button_divide = tk.Button(root, text="Divide", command=divide, bg="#f39c12", fg="#000000")
button_divide.grid(row=3, column=1, padx=5, pady=5)

# Run the main event loop
root.mainloop()
```
